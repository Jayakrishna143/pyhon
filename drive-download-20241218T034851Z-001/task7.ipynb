There are many ways to create Dictionary.


*   most comman way is dict_1 ={K:V}
*   dict Constructor can be used



```python
# methon 1
# common method
dictionary = {"age": "23", "place": "hyderabad"}
print(dictionary)
```

```python
# method 2
# using dict constructor
dictionary = dict(age = "23", place = "hyderabad")
print(dictionary)
```

```python
# method 3
#dict constructor with list of tuples
dictionary = dict([("age","23"),("place","hyderabad")])
print(dictionary)
```

```python
# method 4
# dict constructor with a dictionary comprehension
dictionary = {f"number{i}": f"{i}" for i in range(1,9)}
print(dictionary)
```

```python
# method 5
# dict constructor with list of lists
dictionary = dict([["age","23"],["place","hyderabad"]])
print(dictionary)
```

```python
# method 6
# dict constructor with tuple of tuples
dictionary = dict((("age","23"),("place","hyderabad")))
print(dictionary)
```

```python
# method 7
# using dict constructor with list of key_value pairs from zip
keys = ["age","place"]
values = ['23','hyderabad']
dictionary = dict(zip(keys,values))
print(dictionary)
```

```python
#method 8
# using dict constructor with another Dictionary
original_dict = {"age": "23", "place": "hyderabad"}
dictionary = dict(original_dict)
print(dictionary)
```

```python
# method9
# from dict.fromkeys method
keys = ["age","place"]
dictionary = dict.fromkeys(keys,'200')
print(dictionary)
```

```python
# method10
# using dict.update with a empty dictionary
dictionary = {}
dictionary.update({"age": "23", "place": "hyderabad"})
print(dictionary)
```

```python
# method 11
# using enumerate to create dictionary where keys are index
values = ["age","place"]
dictionary = dict(enumerate(values))
print(dictionary)
```
