```python

methods_with_boolean_output = [
    'isalnum',
 'isalpha',
 'isascii',
 'isdecimal',
 'isdigit',
 'isidentifier',
 'islower',
 'isnumeric',
 'isprintable',
 'isspace',
 'istitle',
 'isupper'
]
```

```python
# Example: str.isalnum()  Checks weather the string contains alphabets and numbers but not spaces
s1 = "Jaya123"
s2 = "Jaya 123"
print(s1.isalnum())
print(s2.isalnum())
```

```python
# Example: str.isalpha() True only if string contain alphabets of capital or smaller case but not numbers
s1 = "Jaya"
s2 = "Jaya123"
print(s1.isalpha())
print(s2.isalpha())
```

```python
# Example: str.isdigit() checks only digits from 0-9 contains
s1 = "12345"
s2 = "12345a"
print(s1.isdigit())
print(s2.isdigit())
```

```python

# Example: str.isdecimal() checks if all characters in a string are decimal characters.
s1 = "12345"
s2 = "@"
print(s1.isdecimal())
print(s2.isdecimal())
```

```python
# Example: str.islower() checks weather alphabets are smaller case or not
s1 = "Jaya"
s2 = "jaya"
print(s1.islower())
print(s2.islower())

```

```python
# Example: str.isupper() checks weather alphabets are capital or not
s1 = "JAYA"
s2 = "jaya"
print(s1.isupper())
print(s2.isupper())
```

```python
# Example: str.istitle()  checks if the first letter of each word is capitalized and the rest are lowercase
s1 = "Hello World"
s2 = "hello world"
print(s1.istitle())
print(s2.istitle())
```

```python
# Example: str.isspace() checks strings whether its only space or it contains any alpha numaric characters
s1 = "   "
s2 = "Hello World"
print(s1.isspace())
print(s2.isspace())
```

```python
# Example: str.isnumeric() it checks if given string contains numbers ,digits from 1-10,fractions,roman numbers etc
s1 = "12345"
s2 = "12345a"
print(s1.isnumeric())
print(s2.isnumeric())
```

```python
# Example: str.isidentifier() it can contain letters and numbers and underscores but should no start with a number
# and also keywords like if else while etc
s1 = "x1"
s2 = "1x"
print(s1.isidentifier())
print(s2.isidentifier())
```

```python
# Example: str.isascii() checks if given string contains ascii characters or not
s1 = "Hello!"
s2 = "ćâŕ"
print(s1.isascii())
print(s2.isascii())
```

```python
# Example: str.isprintable() checks if all characters in a string are printable.
s1 = "Hello, World!"
s2 = "Hello\nWorld"
s3 = "Hello\tWorld"

print(s1.isprintable())  # Output: True
print(s2.isprintable())  # Output: False
print(s3.isprintable()) # Output: True
```

## Dictionary **setdefault** Method


### **dict.setdefault()**
The setdefault() method in dictionaries is used to get the value of a key if it exists.
If the key does not exist, it inserts the key with a default value and returns that value.

**Eg**: dictionary.setdefault(key, default_value)

* key: The key to look for in the dictionary.
* default_value (optional): The value to set if the key doesn't exist. Default - None.




```python

# Example 1: if Key exists in dictionary
dict1 = {"a": 1, "b": 2}
value = dict1.setdefault("a")
print(dict1)
print(value)

# Example 2: if Key does not exist in dictionary
dict1 = {"a": 1, "b": 2}
value = dict1.setdefault("c", 100)
print(dict1)
print(value)
```
