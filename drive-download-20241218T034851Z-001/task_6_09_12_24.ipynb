
# Set Methods

Sets are an essential data structure in Python,
1.   sets are used for storing unique elements.

2.   They are unordered and mutable,



## Creating Sets

 set  can be created by using curly braces **{}** or the **set()**

### Example:


```python

#  sets using curly braces
x = {1, 2, 3, 4}

#an empty set using set() ,{} creates a dictionary
a = set()

print("Set x:", x)
print(f"Empty set a:, {a},{type(a)}")
```

```python
#  update() - Adds elements from another set or iterable to the current set.
# Example
x = {1, 2, 3}
a = {4, 5}
x.update(a)
print("Set after update:", x)
```

```python
# add () adds elements to the set
# Example of add() method
x = {1, 2, 3}
x.add(4)
print(x)
```

```python
# pop() removes element from the set
# Example of pop() method
x = {1, 2, 3}
removed_element = x.pop()
print("Removed element:", removed_element)
print("Set after pop:", x)
```

```python
# removes selected element from the set
# Example of remove() method
x = {1, 2, 3}
x.remove(2)
print("Set after remove:", x)
```

```python
# similar to remove() method but when a element which is not present in set is given no error appears
# Example of discard() method
x = {1, 2, 3}
x.discard(3)
x.discard(4)
print("Set after discard:", x)
```

```python
# clear() is similar to list it clears all the elements in the set
# Example of clear() method
x = {1, 2, 3}
x.clear()
print("Set after clear:", x)
```

```python
# cop
# Example of copy() method
x = {1, 2, 3}
a = x.copy()
print("Original set x:", x)
print("Copied set a:", a)
```


## Shallow Copy vs Deep Copy

A **shallow copy** creates a new object but only copies the reference. Changes in the original object reflect in the shallow copy.

A **deep copy** creates a new Indipendent copy . Changes to the original object do not affect the deep copy.

### Examples:


```python
# shallow copy
import copy

# Shallow copy example
x = [{1, 2}, {3, 4}]
a = copy.copy(x)

# Modify original
x[0].add(5)

print("Original:", x)
print("Shallow copy:", a)  #  changes in the original can be seen
```

```python

# Deep copy example
x = [{1, 2}, {3, 4}]
a = copy.deepcopy(x)

# Modify original
x[0].add(5)

print("Original:", x)
print("Deep copy:", a)  # changes cannot be seen
```
