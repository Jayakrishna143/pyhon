### Order of Execution

*   uses BODMAS OR PEMDAS
In BODMAS the priority while solving a problem starts with brackets,order,division,multiplication,addtion and subtraction  
In PEMDAS it starts with Parentheses, Exponents, Multiplication, Division, Addition, and Subtraction.

so according to this the solution process is


1.  Exponentiation (**) : 5 ** 2 = 25

1.   Multiplication (*) : 2 * 3 = 6
2.   Division (/) : 6 / 4 = 1.5
2. Addition (+) : 10 + 1.5 = 11.5

1.   subtraction (-) : 11.5 - 25 = -13.5


```python
#4
expression = (8 + 2) * (3 - 1) ** 2 / 2
print(expression)
```

1.   As explained above the Process of PEMDAS is followed

1.   Parentheses override the default operator process and ensures that the problem is solved from inside then continues with other
3.   If parentheses are not used then the problem would first start with power values then the solution would be totally different
