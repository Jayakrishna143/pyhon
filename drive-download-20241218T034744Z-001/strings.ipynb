```python
''' make sure stirng contains alphanumaric and special characters '''
x = input('Enter a string: ')
a,b,c = [],[],[]
for i in x:
  if i.isalpha():
    a.append(i)
  elif i.isdigit():
    b.append(i)
  else:
    c.append(i)
print(a,b,c)


```

The default values of start and end of range function is 0 and 1 respectiverly

```python
Y = input('enter any string only alphabets: ')
a,b = [],[]
for i in Y:
  if ord(i) in range(65,91):
    a.append(i)
  elif ord(i) in range(97,123):
    b.append(i)
  else:
    print('learn to read ')

print(a,b)
```

replicate .lower method using an input statement empty string a for loop conditional statement range function

```python
x = input('enter a string')
e_string = ''
for i in x:
  if ord(i) in range(65,91):
    e_string += chr(ord(i)+32)
  elif ord(i) in range(97,123):
    e_string += i
  else:
    print('only alphabets please')
print(x,e_string)
```
